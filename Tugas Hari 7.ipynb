<h2>Soal 2: Filter DataFrame Dengan Boolean</h2>

1. Tampilkan semua data siswa yang level pendidikan orang tuanya 'high school' dan 'bachelor degree'
2. Tampilkan semua data siswa yang mempunyai nilai lebih dari 50 dengan test preparation course 'completed'
3. Tampilkan semua data siswa yang mempunyai nilai lebih dari 70, berjenis kelamin perempuan dengan test preparation course 'completed'

```python
# saya bermasalah dengan nama kolom yang dipisahkan dengan spasi
# maka saya mengganti nama kolom yang dipisahkan dengan spasi
# menjadi tidak menggunakan spasi

df.rename(columns={'parental level of education':'parental_level_of_education'}, inplace=True)
df.rename(columns={'test preparation course':'test_preparation_course'}, inplace=True)
df
```

```python
# code here
# Tampilkan semua data siswa yang level pendidikan orang tuanya 'high school' dan 'bachelor degree'
hs = df.parental_level_of_education == 'high school'
bd = df.parental_level_of_education == 'bachelor degree'

df[(hs) | (bd)]
```

```python
# Tampilkan semua data siswa yang mempunyai nilai lebih dari 50 dengan test preparation course 'completed'
completed = df.test_preparation_course == 'completed'
nilai = df.nilai > 50

df[(completed) & (nilai)]
```

```python
# Tampilkan semua data siswa yang mempunyai nilai lebih dari 70,
# berjenis kelamin perempuan
# dengan test preparation course 'completed'

nilai = df.nilai >70
sex = df.gender == 'female'
course = df.test_preparation_course == 'completed'

df[(nilai)&(sex)&(course)]

```

<h2>Soal 3: Data Transformation</h2>

lakukan data transformasi terhadap column 'nilai' dengan fungsi transformasi, berikut rumus dari normalisasi

![alt text](https://drive.google.com/uc?id=1wed-EbRSTA0JP1ceDp6lcKBqH2yJ4aa3)

```python
#code here

def normal(x):
  return (x-x.min())/(x.max()-x.min())

df[['nilai']].apply(normal)
```
