# Soal 2. Persiapan Data

*   Hapus nilai missing dalam rows dataset menggunakan atribut.dropna()

*   Lakukan grouping data berdasarkan kolom 'species' dengan menggunakan agregasi mean()



```python
#code here
# Hapus nilai missing dalam rows dataset menggunakan atribut.dropna()
df2 = df.dropna()
print(df2)
```

```python
Spesies = df2.groupby('species').agg({
    'bill_length_mm': 'mean',
    'bill_depth_mm': 'mean',
    'flipper_length_mm' : 'mean',
    'body_mass_g' : 'mean'
    })
print(Spesies)
```

# Soal 3. Univariate Analysis
Buatlah sebuah analisa univariate dengan menggunakan perintah seperti berikut:

*   Buatlah sebuah Pie plot dari dataset hasil grouping untuk kolom flipper_length_mm seperti expected output berikut:
*   Buatlah sebuah plot histogram dari dataset awal (sebelum grouping) untuk kolom body_mass_g
*   Jelaskan insight yang diperoleh dari ke dua grafik tersebut.





```python
#code here
# pie chart flipper_length_mm

fig, ax = plt.subplots()
ax.pie(Spesies['flipper_length_mm'], autopct='%1.1f%%', labels=Spesies.index)
```

Expected Output:

```python
plt.hist(df2['body_mass_g'])
plt.xlabel('Body_mass_g')
plt.ylabel('Count')
```

Analisa:


Pie chart ini menyatakan bahwa Spesies Gentoo memiliki rata-rata panjang sirip (flipper length) yang terbesar dibandingkan dengan yang lain. Presentse nya mencapai 36% dari keseluruhan.


Histogram ini menyatakan sebaran massa tubuh dari semua spesies penguin. Spesies yang memiliki massa tubuh antara 3500 sampai 4000 gram memiliki jumlah yang paling banyak dibandingkan sebaran massa tubuh yang lain.

# Soal 4. Bivariate Analysis



*   Buatlah sebuah grafik catplot menggunakan seaborn dengan x='sex', y='bill_length_mm' dan kind 'bar' seperti expected output di bawah
*   Buatlah sebuah grafik scatter plot menggunakan seaborn dengan x='body_mass_g', y='flipper_length_mm', hue='bill_length_mm' seperti expected dibawah
*   Jelaskan insight yang diperoleh dari hasil analisa bivariate tersebut






```python
#code here

# catplot
palette = sns.color_palette()
sns.catplot(data=df2, x='sex', y='bill_length_mm', kind='bar', palette=palette)
```

Expected Output:

Catplot

![](https://drive.google.com/uc?id=1imCbGD9cWqXUjd33tAh6D0OH20vVpWea)

Scatter plot

![](https://drive.google.com/uc?id=1_k734IpRYq6Q1wF57hwzSEjnVd2Sii2Z)

```python
# scatter plot

sns.scatterplot(data=df2, x='body_mass_g', y='flipper_length_mm', hue='bill_length_mm')
```

Analisa:

Catplot diatas menerangkan bahwa penguin jantan di semua spesies memiliki panjang paruh yang lebih panjang dibandingkan penguin betina.


Scatter plot diatas menerangkan korelasi/hubungan antara massa tubuh dan panjang sirip penguin pada semua spesies.
Terlihat bahwa semakin panjang siripnya, maka besar pula massa nya (korelasinya positif). Warna scatter plot juga berdasarkan panjang paruhnya, secara umum terlihat juga berkorelasi positif dengan massa dan panjang sirip.

Analisa:
Pairplot memberikan gambaran visual yang komprehensif terkait data. Kita bisa memahami korelasi antara variabel yang ada.

Pada diagonal utama, histogram yang ditampilkan untuk melihat distribusi atau sebaran data pada masing-masing variabel.

Pada scatter plot yang berada di luar diagonal menunjukkan korelari/hubungan antar variabel. Kita bisa melihat pola yang dihasilkan, apakah berkorelasi positif, negatif, atau malah tidak sama sekali.

Selain itu, kita juga bisa menjelaskan jika kumpulan scatter (cluster) terbagi atas 2 kumpulan atau lebih.  Misalnya korelasi antara massa dan kedalaman paruh pinguin, terdapat 2 cluster, yaitu penguin  bermassa besar akan memiliki kedalaman paruh yang relatif kecil sedangkan penguin bermassa kecil memiliki paruh yang relatif sangat dalam. Pada scatter plot juga bisa melihat data yang outlier.
