<h2>Soal 3: If-Else Statement</h2>

Lengkapi kode untuk menghasilkan suatu output yang di harapkan

- Bualah sebuah if-else statement yang dimana akan mem-print 'High' jika grade adalah 'A' dan price lebih dari 100000, kemudian mem-print 'Medium' jika grade adalah 'A' dan price lebih dari 50000 dan memprint 'low' jika grade adalah 'A' dan price lebih kecil dan sama dengan 50000.

```python
#code here
grade = 'A'
price = 150000

if grade == 'A' and price > 100000:
  print('Hight')
elif grade == 'A' and price > 50000:
  print('Medium')
elif grade == 'A' and price <= 50000:
  print('Low')
```

<h2>Soal 4: Comparison Operator dengan fungsi</h2>

Buatlah sebuah fungsi yang menerima satu argument bertipe data numeric dan menghasilkan sebuah return sebagai berikut :
- menghasilkan return 'Aneh' jika nilai dari argument tersebut adalah bilangan ganjil
- menghasilkan return 'Tidak Aneh' jika nilai dari argument tersebut adalah bilangan genap dan diantara nilai 2 sampai 5 (2 dan 5 termasuk)
- menghasilkan return 'Aneh' jika nilai dari argument tersebut adalah bilangan genap dan diantara nilai 6 sampai 20 (6 dan 20 termasuk)
- menghasilkan return 'Tidak Aneh' jika nilai dari argument tersebut adalah bilangan genap dan lebih besari dari 20

```python
# Buat fungsi
def Fungsi(x):
  if x%2==0:
    if (x>=2 and x<=5) or x>20:
      return 'Tidak aneh'
    elif x>=6 and x<=20:
      return "Aneh"
  else:
    return "Aneh"

print(Fungsi(22))
```

<h2>Soal 5: While Loop dan For Loop</h2>

- Apa perbedaan while loop dan for loop?
- Berikan contoh sederhana cara menggunakan while loop dan for loop

Perbedaan *while loop* dan *for loop* adalah *while loop* mengulangi satu blok kode hingga kondisinya salah, sedangkan  *for loop* mengulangi satu blok kode sebanyak beberapa kali tertentu.


```python
# jawaban contoh

# while loop
x=5
while x<30:
  print(x, end=' ') #looping terus berjalan jika kondisinya benar
  x+=1 # jika ini tidak ada, maka looping akan terus berjalan

  # looping akan terhenti ketika kondisinya salah.
  # pada kode ini looping akan terhenti ketika x nya sudah tidak kurang dari 30

print('')
print('='*70)

# for loop
for x in range(0,10):
  print(x, end=' ')
```

```python
# a = 10
a = -10


def fungsi_while(angka):
  while angka != 0:
    print(angka)
    if angka <0:
      angka +=1
    else:
      angka -=1

fungsi_while(a)
```

<h2>Soal 7: For Loop</h2>

Lengkapi kode untuk menghasilkan suatu output yang diharapkan:





*   Cari siswa mana saja yang memiliki nilai lebih dari sama dengan 80. Masukkan kedalam sebuah list. print hasilnya






```python
data_siswa = [
    {
        "nama":"Budi",
        "nilai": 90
    },
    {
        "nama":"Nina",
        "nilai": 78
    },
    {
        "nama":"Rudi",
        "nilai": 91
    },
    {
        "nama":"Olivia",
        "nilai": 76
    },
    {
        "nama":"Leo",
        "nilai": 80
    },
    {
        "nama":"Liam",
        "nilai": 67
    },
    {
        "nama":"Sheila",
        "nilai": 76
    }
]
```

```python
#code here
list_baru = []
for data in data_siswa:
  if data['nilai'] >=80:
    list_baru.append(data['nama'])

print(list_baru)
```

Expected output:

['Budi', 'Rudi', 'Leo']
