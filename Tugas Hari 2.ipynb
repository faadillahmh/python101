# Data Structure

<h2>Soal 1: Perbedaan Data Structure</h2>

Jawab Pertanyaan di bawah ini:

Jelaskan perbedaan dari List, Tuple, Set dan Dictionary!

**Perbedaan Dasar dari List, Tuple, Set, dan Dictonary**
*   **List** merupakan kumpulan data yang dibuat secara berurutan dan dinyatakan dengan ekspresi menggunakan tanda kurung siku.
*   **Tuple** merupakan kumpulan data yang dibuat secara berurutan dan dinyatakan dengan ekspresi menggunakan tanda kurung.
*   **Set** merupakan kumpulan data yang tidak berurutan dan dinyatakan dengan ekspresi menggunakan kurung kurawal
*   **Dictionary** merupkan kumpulan data yang tak berurutan yang memetakan antara *key* dan *value* dari suatu data.
Dictionary diekpresikan menggunakan kurung kurawal (dengan elemen *key* dan *value*).

## Soal 6: List Comprehension

Gunakan metode **list comprehension** untuk mencari anggota dari S yang habis di bagi 2, kemudian assign hasilnya dalam bentuk list ke dalam variabel T.

```python
S = [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

T = [i for i in S if i%2==0]

print(T)
```

Expected Output:

[0, 4, 16, 36, 64]

## Soal 12: Remove Duplicate using set


Hilangkan nilai duplikat dari sebuah objek list dengan menggunakan cara set sehingga menjadi sebuah tipe set

```python
obj_list = [1, 2, 3, 4, 5, 5, 6, 7, 8, 9, 9, 10]

#using set
set(obj_list)
```

## Soal 13: Mengubah dan menghapus anggota set

Ubahlah dan hapus anggota set sehingga mendapatkan hasil yang diinginkan


```python
set_1 = {1, 2, 3, 4, 5}

#tambahkan anggota set dengan nilai {6,7,8}
set_1.update([6,7,8])

#hapus nilai anggota set 8
set_1.remove(8)
print(set_1)
```

Expected output:

{1, 2, 3, 4, 5, 6, 7}

## Soal 14: Operasi pada Set

Carilah irisan dari ke dua set dengan menggunakan metode intersection

```python
set_2 = {6, 8, 9, 10, 24}
set_3 = {6, 10, 8, 25, 13}

#intersection set
print(set_2 & set_3)
```

Expected output:


{6, 8, 10}
