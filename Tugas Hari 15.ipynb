```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset('diamonds')
df
```

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

```python
interval=df['interval']=pd.qcut(df['depth'], q=6)
interval
```

```python
sns.set({'figure.figsize':(13,13)})
palette = "ch:r=-.2,d=.3_r"

#code here
bins = [42.9, 60.6, 61.4, 61.8, 62.3, 62.9, 79.0]
order = ['48', '54', '60', '66', '72','78']
df['depth_'] = pd.cut(df['depth'], bins=bins, labels=order)

sns.scatterplot(data=df, x='carat', y='price', hue='clarity', style='depth_', palette = "ch:r=-.2,d=.3_r", style_order=order, size='depth_', sizes=(30, 10), markers='o')
```

# Soal 2: Catplot

Buatlah sebuah Categorical plot dengan kind = 'boxplot', sehingga diperoleh hasil seperti expected output berikut:


```python
df.head()
```

```python
fig, axes = plt.subplots(1, 5, figsize=(20, 6), sharey=True)

cuts = df['cut'].unique()

for ax, cut in zip(axes, cuts):
    sns.boxplot(x='clarity', y='price', data=df[df['cut'] == cut], palette='Set2', ax=ax)
    ax.set_title(f'Cut = {cut}')
    ax.set_xlabel('Clarity')
    ax.set_ylabel('Price')
    ax.set_ylim

plt.tight_layout()

plt.show()
```

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price serta markers =  ['p', 'o', 's', 'D', 'X'] sehingga dihasilkan output sesuai expected.

```python
df = df.sample(n=300, random_state=123)

#code here

markers = ['p', 'o', 's', 'D', 'X']
g = sns.pairplot(df, vars=['carat', 'depth', 'table', 'price'], hue='cut', markers=markers)
plt.show()
```
