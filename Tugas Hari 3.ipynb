<h2>Soal 2: Built-in Function</h2>

Jawab Pertanyaan di bawah ini:
    
Apa itu built-in Function?

Sebutkan 3 built-in Function di pyhon!

Berikan contoh masing2 1 cara penggunaan built-in function yang kamu sebutkan sebelumnya

Built-in Function merupakan sebutan untuk fungsi bawaan yang sudah ada di Python. Setidaknya, ada 68 built-in function yang terdapat pada Python versi terbaru. Built-ini function yang sering dipakai adalah fungsi print().

3 contoh built-in function di Python
*   Fungsi abs() --> mengembalikan nilai absolut dari suatu bilangan
*   Fungsi bin() --> mengembalikan representasi biner dari integer tertentu
*   Fungsi pow() --> menghasilkan nilai pangkat dari suatu bilangan








```python
# Contoh penggunaan built in function

# fungsi print()
# print(yang akan ditampilkan)
print('Halo')

# fungsi abs()
# abs(bilangan)
abs(-2.9)

# fungsi bin()
# bin(bilangan integer)
bin(4)

# fungsi pow()
# pow(bilangan, bilangan pangkat)
pow(3,2) # 3 pangkat 2
```

<h2>Soal 3: Method dan Function</h2>

Jawab Pertanyaan di bawah ini:
    
- Apa perbedaan method dan function?



1. Metode merupakan tindakan/operasi yang bisa digunakan/dipanggil berdasarkan jenis objek tertentu pada suatu bahasa pemrograman. Misalnya pada objek list di Python, memiliki banyak metode yang digunakan seperti append() untuk menambahkan daftar list.


2. Function merupakan kode yang dapat digunakan untuk melakukan tugas tertentu. Fungsi dapat mengambil argumen (input), melakukan operasi atau manipulasi pada argumen tersebut, dan mengembalikan nilai (output) yang diinginkan. Fungsi dapat digunakan berkali-kali dalam program.
Contoh yang paling umum kita dapat mendefinisikan sebuah fungsi dengan menggunakan syntax def, kemudian memberikan nama pada fungsi tersebut dan menentukan argumen yang diperlukan. Fungsi tersebut dapat melakukan operasi tertentu dan mengembalikan hasil menggunakan return.

<h2>Soal 5: Membuat Simple Function</h2>

Buatlah suatu fungsi yang menerima satu input argumen berbentuk list dan mempunyai elemen bertipe numeric semua, dimana fungsi tersebut berguna untuk menghitung rata2 dari kumpulan elemen list tersebut. namai fungsi tersebut 'mean_list'

```python
obj_list = [11.25, 18.0, 20.0, 10.75, 9.50, 13.45, 23.0, 7.0, 8.45, 2.50]
def mean_list(inp_list):
  return sum(obj_list) / len(obj_list)

print(mean_list(obj_list))
```

Expected Output:

12.39
