<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
#code here

fig, ((ax1, ax2, ax3)) = plt.subplots(1, 3, sharex=True, sharey=True, figsize=(14,8))
ax1.plot(x,tan, label='tanX', color='blue')
ax1.set_title('Grafik tanx')
ax1.set_xlim(-8,8)
ax1.set_xticks(range(-6,8,2))
ax2.plot(x,sin, label='sinX', color='red')
ax2.set_title('Grafik sinx')
ax2.set_xlim(-8,8)
ax2.set_xticks(range(-6,8,2))
ax3.plot(x,cos, label='cosX', color='green')
ax3.set_title('Grafik cosx')
ax3.set_xlim(-8,8)
ax3.set_xticks(range(-6,8,2))
fig.legend(loc='upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here

fig, ax = plt.subplots(figsize=(8,7))
ax.plot(x, y)
ax.set_title('Grafik CosX')

ins1=ax.inset_axes([0.9, 0.9, 0.4, 0.3])
ins1.plot(x,y4, color='black')
ins1.plot(x,y)
ins1.set_title('Grafik CosX^4')
ins1.set_xlim(-8,8)
ins1.set_xticks(range(-6,8,2))


ins2=ax.inset_axes([0.9, -0.1, 0.4, 0.3])
ins2.plot(x,y3, color='green')
ins2.plot(x,y)
ins2.set_title('Grafik CosX^3')
ins2.set_xlim(-8,8)
ins2.set_xticks(range(-6,8,2))

ins3=ax.inset_axes([-0.3, -0.1, 0.4, 0.3])
ins3.plot(x,y2, color='red')
ins3.plot(x,y)
ins3.set_title('Grafik CosX^2')
ins3.set_xlim(-8,8)
ins3.set_xticks(range(-6,8,2))

ins4=ax.inset_axes([-0.3, 0.9, 0.4, 0.3])
ins4.plot(x,y5, color='yellow')
ins4.plot(x,y)
ins4.set_title('Grafik CosX^5')
ins4.set_xlim(-8,8)
ins4.set_xticks(range(-6,8,2))
```
